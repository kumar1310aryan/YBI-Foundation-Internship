# Ice-cream Revenue Prediction

```python
import pandas as pd
```

```python
icecream = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Ice%20Cream.csv')
```

```python
icecream.to_csv('icecream.csv',index=False)
```

```python
icecream.head()
```

```python
df=icecream
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature'], df['Revenue'], color='blue')
plt.title('Temperature vs Revenue')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.show()
```

```python
icecream.info()
```

```python
icecream.describe()
```

```python
icecream.columns
```

```python
y=icecream['Revenue']
```

```python
X=icecream[['Temperature']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Regression Line')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Testing Data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.title('Temperature vs Revenue with Regression Line')
plt.xlabel('Temperature')
plt.ylabel('Revenue Generated')
plt.legend()
plt.show()
```

## Prediction

```python
Temperature = pd.DataFrame({'Temperature': [24.6]})
predicted_score = model.predict(Temperature)
print(f"Predicted Revenue : {predicted_score[0]}")
```

So the revenue generated is 572 according to our model.
